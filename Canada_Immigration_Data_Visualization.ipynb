## 1. Immigration from Haiti

**Back drop**

In 2010, Haiti suffered a catastrophic magnitude 7.0 earthquake.

The quake caused widespread devastation and loss of life and about three million people were affected by this natural disaster.

As part of Canada's humanitarian effort, the Government of Canada stepped up its effort in accepting refugees from Haiti.

```python
# passing in years 1980 - 2013 and excluding the 'total' column
haiti = df.loc['Haiti',years] 
haiti.index = haiti.index.astype(int)

# plotting a line graph
haiti.plot(kind='line', figsize=(10,7))

plt.title('\n Immigration from Haiti \n', fontdict={'fontsize' : 18})
plt.ylabel('\n Number of immigrants \n', fontdict={'fontsize' : 14})
plt.xlabel('\n Year \n', fontdict={'fontsize' : 14})

# setting tick values
ticks = range(1980, 2016, 5)
plt.xticks(ticks = ticks, fontsize=10)
plt.yticks(fontsize=10)

# annotate the 2010 Earthquake. 
plt.annotate('2010 Earthquake', xy=(2003,6300)) 

# setting xlimit, ylimit values
plt.xlim(1978,2017)
plt.ylim(800,7200)

plt.show() 
plt.close()
```

**Inference**:<br>
We can clearly see a spike in immigrants from 2010, from Haiti. Reemphasizing the point presented in the case background earlier.

## 2. Immigration Trend of Top 5 Countries

```python
# Filtering out Top 5 immagration values into a separate DataFrame
df_top5 = df.sort_values(by='Total', ascending=False).head(5)
# Transposing the dataframe inorder to adapt for the use case
df_top5 = df_top5[years].transpose()
# To check if the dataframe is correct
# print(df_top5)

# changing the index values of df_top5 to type integer for plotting
df_top5.index = df_top5.index.astype(int)

# Plotting the line Graph
df_top5.plot(kind='line', figsize=(15, 10))

plt.title('\n Immigration Trend of Top 5 Countries \n', fontdict={'fontsize' : 18})
plt.ylabel('\n Number of Immigrants \n', fontdict={'fontsize' : 14})
plt.xlabel('\n Years \n', fontdict={'fontsize' : 14})

# setting the tick values
xticks = range(1980, 2016, 5)
yticks = range(0, 50000, 5000)
plt.xticks(ticks=xticks, fontsize=10)
plt.yticks(ticks=yticks, fontsize=10)

# setting x and y limits
plt.xlim(1978,2017)
plt.ylim(-1000,50000)

# Add legend as a solid block
plt.legend(title='Country', loc='upper left', frameon=True)

plt.show()
plt.close()
```

**Inference**:
1. We can see **India & China** contribute the **most** of the immigrants.
2. Also we can observe that, India & China represented by Blue & Orange respectively, have **similar pattern** of immigration.
3. We see a **spike** in immigrants from **UK** during the year 1994 and then a gradual declining trend after 1996.
4. There is a sudden spike in Immigrants from **Philipines** after 2004

## 3. Immigration Trend from India & China

## 8a. Icelandic Immigrants to Canada from 1980 to 2013

**Backdrop**

The 2008 - 2011 Icelandic Financial Crisis was a major economic and political event in Iceland.

Relative to the size of its economy, Iceland's systemic banking collapse was the largest experienced by any country in economic history.<br>
The crisis led to a severe economic depression in 2008 - 2011 and significant political unrest.

```python
# Dataframe containing icelandic data
df_iceland = df.loc['Iceland', years]

# plotting a bar graph
df_iceland.plot(kind='bar', color='steelblue', figsize=(10, 7))

plt.title('\n Icelandic Immigrants to Canada from 1980 to 2013 \n', fontdict={'fontsize' : 18})
plt.xlabel('\n Year \n', fontdict={'fontsize' : 14})
plt.ylabel('\n Number of Immigrants \n', fontdict={'fontsize' : 14})

# xticks
plt.xticks(rotation=90, fontsize=10)

# Annotate arrow
plt.annotate('',  # s: str. blank for no text
             xy=(32, 70),  # head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),  # base of the arrow at point (year 2008 , pop 20)
             xycoords='data',  # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis',  # text to display
             xy=(28, 30),  # start the text at at point (year 2008 , pop 30)
             rotation=72.5,  # arrow inclination
             va='bottom',  # text to be vertically 'bottom' aligned
             ha='left',  # text to be horizontally 'left' algned.
             )

plt.show()
plt.close()
```

**Inference**:

## 9a. Immigration to Canada by Continents [1980 - 2013]

```python
# Creating a dataframe using groupby() to contain summed data grouped by Continents
df_continents = df.groupby('Continent').sum()


colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge.

# plot a pie graph
df_continents['Total'].plot(kind='pie',
                            figsize=(15, 7),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90° (Africa)
                            shadow=True,       # add shadow
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.1,    # to show percentages generated by autopct, outside 
                            colors=colors_list,  # add custom colors
                            explode=explode_list # 'explode' closely fitting wedges for better visibility
                            )

# scale the title up by 12% to match pctdistance
plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12, fontdict={'fontsize' : 18}) 

plt.axis('equal') 

# add legend
plt.legend(labels=df_continents.index, loc='upper left', frameon=True) 

plt.show()
plt.close()
```

**Inference**:
1. The figure shows the contribution from each of the continents towards the total immigration during the period 1980 - 2013.
2. We can see **Asia** contributes a majority share - 51.8%
3. Followed by **Europe** - 22% and **Latin America** - 11.9%
